ìš°ì¸¡ì— ìˆëŠ” "Most Popular" ì„¹ì…˜ ê°€ìš´ë°ì—ì„œ í•œ ë°ì´í„°ì…‹ì„ ì„ íƒí•œ ë’¤, í•´ë‹¹ ë°ì´í„°ì…‹ì„ URLë¡œ ë¶€í„° ë°”ë¡œ ì½ì–´ë³´ì„¸ìš”. (ì‘ì—… ì½”ë“œë¥¼ ì‘ì„±í•˜ë©°, ì•„ë˜ì˜ ì§ˆë¬¸ì— ëŒ€í•´ ëŒ€ë‹µí•  ìˆ˜ ìˆì–´ì•¼í•©ë‹ˆë‹¤.) 

- í—¤ë”ê°€ ì •ìƒì ìœ¼ë¡œ ë³´ì—¬ì§€ë‚˜ìš”?
- ì²« 5ì¤„ í˜¹ì€ ë§ˆì§€ë§‰ 5ì¤„ì„ í™•ì¸í•˜ê³  ê°ê° ìˆœì„œë¥¼ ê°€ì§€ê³  ì •ë ¬ì´ ë˜ì–´ìˆë‚˜ìš”?

```python
# 1) UCI ë°ì´í„°ì…‹ ë¶ˆëŸ¬ì˜¤ê¸° (URLì„ í†µí•´)
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.data'

import pandas as pd

df1 = pd.read_csv(data_url)
df1.columns = ['animal name', 'hair', 'feathers', 'eggs', 'milk', 'airborne', 'aquatic', 
              'predator', 'toothed', 'backbone', 'breathes', 'venomous', 'fins', 'legs', 'tail', 'domestic', 'catsize', 'type']
print(df1)

# í—¤ë”ê°€ ì •ìƒì ìœ¼ë¡œ ë³´ì—¬ì§€ë‚˜ìš”?
print(list(df1.columns.values))
print('---')
# ì²« 5ì¤„ í˜¹ì€ ë§ˆì§€ë§‰ 5ì¤„ì„ í™•ì¸í•˜ê³  ê°ê° ìˆœì„œë¥¼ ê°€ì§€ê³  ì •ë ¬ì´ ë˜ì–´ìˆë‚˜ìš”?
  # ë„¤
print(df1.head())
print('---')
print(df1.tail())
print('---')

# UCI í™ˆí˜ì´ì§€ì—ì„œ ì„¤ëª…í•œ ê²ƒê³¼ ë™ì¼í•œ í–‰ê³¼ ì—´ì„ ê°€ì§€ê³  ìˆë‚˜ìš”?
  # UCI says   Number of Instances: 101
  #            Number of Attributes: 17
  # df.shape = (100, 18)ê°€ ë‚˜ì˜µë‹ˆë‹¤.
print(df1.shape)
print('---')

# ë°ì´í„°ì…‹ì— ê²°ì¸¡ì¹˜ê°€ ì¡´ì¬ í•˜ë‚˜ìš”? (UCIì˜ ì„¤ëª…ê³¼ ì¼ì¹˜ í•˜ëŠ”ì§€)
  # ì•„ë‹ˆì˜¤ (UCI says : Missing Values? No) 
  # UCI ì„¤ëª…ê³¼ ì¼ì¹˜í•©ë‹ˆë‹¤.
print(df1.isnull().all())
print('---')

# NaN í˜¹ì€ ë‹¤ë¥¸ ê²°ì¸¡ì¹˜ë¥¼ ë‹¤ë£¨ëŠ” ì‘ì—…ì„ df.isnull().sum()ì™¸ì˜ ë°©ë²•ìœ¼ë¡œ í•  ìˆ˜ ìˆë‚˜ìš”? (ì˜ˆë¥¼ ë“¤ë©´, np.NaNìœ¼ë¡œ í‘œê¸°ëœ ê°’ì„ ê²°ì¸¡ì¹˜ë¡œ ì²˜ë¦¬í•´ì•¼ í•©ë‹ˆë‹¤)
## ê²°ì¸¡ì¹˜ê°€ ì¡´ì¬í•œë‹¤ëŠ” ê°€ì • í•˜ ì‘ì„±í•˜ì˜€ìŠµë‹ˆë‹¤.
print(df1.fillna(0)) # ëª¨ë“  NaNì„ 0ìœ¼ë¡œ ì¹˜í™˜
print('---')
print(df1.dropna()) # NaNì„ í•˜ë‚˜ë¼ë„ í¬í•¨í•˜ëŠ” í–‰ ì‚­ì œ(default : how='any', axis=0)
print('---')
print(df1.dropna(how='all')) # í–‰ ì „ì²´ê°€ NaNì¸ í–‰ë§Œ ì‚­ì œ
print('---')

# .describe() í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬, ì—¬ëŸ¬ í†µê³„ì¹˜ë¥¼ í™•ì¸í•  ìˆ˜ ìˆë‚˜ìš”?
print(df1.describe())
```
